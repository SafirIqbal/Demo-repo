```python
from sklearn.datasets import make_circles
n_samples = 1000
X,y = make_circles(n_samples, noise = 0.03, random_state = 42)
X,y
```

```python
circles = pd.DataFrame({"X0": X[:,0], "X1" : X[:, 1], "Labels" : y})
circles
tf.constant(X)
tf.constant(y)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.RdYlBu);
```

```python
model_categorical = tf.keras.Sequential([
    tf.keras.layers.Dense(4, activation = "relu"),
    tf.keras.layers.Dense(4, activation = "relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
    
])

model_categorical.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                         optimizer=tf.keras.optimizers.Adam(lr=0.01),
                         metrics=["accuracy"])
history=model_categorical.fit(X,y,epochs=250)
```

```python
model_categorical.evaluate(X,y)
```

```python
def plot_decision_boundary(model,X,y):
    x_min,x_max = X[:,0].min()-0.1,X[:,0].max()+0.1
    y_min,y_max = X[:,1].min()-0.1,X[:,1].max()+0.1
    xx,yy = np.meshgrid(np.linspace(x_min,x_max,100),
                        np.linspace(y_min,y_max,100))
    x_in = np.c_[xx.ravel(),yy.ravel()]
    y_pred = model.predict(x_in)
    
    if len(y_pred[0])>1:
        print("Doing multiclass classification")
        y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
    else:
        print("Doing binary classification")
        y_pred = np.round(y_pred).reshape(xx.shape)
    
    plt.contourf(xx,yy,y_pred,cmap=plt.cm.RdYlBu,alpha=0.7)
    plt.scatter(X[:,0],X[:,1],c=y,s=40,cmap=plt.cm.RdYlBu)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    
    
plot_decision_boundary(model_categorical,X,y)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
model_categorical.fit(X_train,y_train,epochs=250)
```

```python
model_categorical.predict(X_test)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.title("Train")
plot_decision_boundary(model_categorical,X_train,y_train)
plt.subplot(1,2,2)
plt.title("Test")
plot_decision_boundary(model_categorical,X_test,y_test)
plt.show();
```
